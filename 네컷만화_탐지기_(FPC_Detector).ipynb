```python
# Step 1: FPC 모델사용에 필요한 라이브러리 등 설치

import os
import subprocess
from google.colab import files
import gdown
import zipfile
import shutil

# Download weights file from Google Drive
!git clone https://github.com/ultralytics/yolov5.git
!pip install -U git+https://github.com/ultralytics/yolov5.git
!pip install -r /content/yolov5/requirements.txt

weights_url = 'https://drive.google.com/uc?id=12HYfiGvOFYj3cbCj-iHxvCmAi-AgV0xs'
weights_path = '/content/FPC-weight.pt'
gdown.download(weights_url, weights_path, quiet=False)
```

```python
# Step 2: Zip 이미지 폴더를 업로드 하면 네컷만화 탐지기 딥러닝 모델이 네컷만화를 탐지하는 과정을 수행함.

def main():
    print("Four Panel Cartoon Detector (FPC Detector 딥러닝 모델입니다)")
    print("Upload a zip folder containing images (Zip 폴더를 업로드 해주세요)")

    # Create the 'uploads' directory if it doesn't exist
    os.makedirs('uploads', exist_ok=True)

    uploaded_files = files.upload()
    if uploaded_files:
        folder_name = list(uploaded_files.keys())[0]

        # Move the uploaded folder to the 'uploads' directory
        full_folder_path = os.path.join(os.getcwd(), folder_name)
        shutil.move(full_folder_path, os.path.join(os.getcwd(), 'uploads', folder_name))

        # Unzip the uploaded file
        extract_path = os.path.join('uploads', os.path.splitext(folder_name)[0])
        with zipfile.ZipFile(os.path.join('uploads', folder_name), 'r') as zip_ref:
            zip_ref.extractall(extract_path)

        print("Folder uploaded.")
        print("Detecting...")

        # Navigate to the yolov5 directory
        os.chdir('/content/yolov5')

        # Make detect.py executable
        subprocess.run(['chmod', '+x', 'detect.py'])

        # Run object detection using the YOLOv5 script
        weights_path = '/content/FPC-weight.pt'
        source_path = '/content/' + extract_path
        output_path = '/content/Output_folder'

        # Change directory back to the root content directory
        os.chdir('/content')

        # Run object detection using the YOLOv5 script
        command = f"python yolov5/detect.py --weights '{weights_path}' --source '{source_path}' --conf 0.6 --save-txt --project '{output_path}' --name result_images"

        # Execute the command and capture the output and error messages
        completed_process = subprocess.run(command, shell=True, capture_output=True)

        # Print the output and error messages
        print("Output:", completed_process.stdout.decode())
        print("Error:", completed_process.stderr.decode())

        if completed_process.returncode == 0:
            print("Detection finished. Check the output directory.")
        else:
            print("Error occurred during detection.")

if __name__ == '__main__':
    main()

```

```python
# Step 3: 네컷만화 딥러닝 탐지 모델이 탐지한 파일이 text 파일로 저장되는데 이 text 파일을 취합하는 작업을 수행하는 스크립트.
# 이 네컷만화의 파일 이름이 들어있는 text 파일과 전체 이미지 파일이 들어있는 폴더를 비교해서 네컷만화가 있다고 딥러닝 모델이 탐지한 파일만을
# shutil.copy()를 사용하여 이미지를 복사하여 추출하는 스크립트.
# 이때 네컷만화 탐지 모델이 탐지한 최종 파일은 /content/Output_folder/detected_true_images 경로에 저장된 것을 확인 할 수 있다.
# 또한 사용자의 편의를 위해서 사용자의 Local 컴퓨터에 "FPC_model_detected.images.zip" 이름의 압축 폴더로 저장이 된다.

import os
import shutil

output_path = '/content/Output_folder'

# Read the list of file names from the text file
file_names_path = os.path.join(output_path, 'result_images', 'labels')
file_names = os.listdir(file_names_path)

# Define the source and destination folders
source_folder = os.path.join(output_path, 'result_images')
destination_folder = os.path.join(output_path, 'FPC_model_detected_images')

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Iterate through the file names, and if the file exists in the source folder, copy it to the destination folder
for file_name in file_names:
    # Remove the '.txt' extension if present
    file_name = os.path.splitext(file_name)[0]

    # Append the .jpg extension if it's missing
    if not file_name.lower().endswith('.jpg'):
        file_name = f'{file_name}.jpg'

    source_file = os.path.join(source_folder, file_name)
    destination_file = os.path.join(destination_folder, file_name)

    if os.path.exists(source_file):
        shutil.copy(source_file, destination_file)
        print(f'Copied {file_name} to {destination_folder}.')
    else:
        print(f'File {file_name} not found in {source_folder}')

# Zip the destination folder
shutil.make_archive('/content/FPC_model_detected_images', 'zip', destination_folder)

# Download the zip file
from google.colab import files
files.download('/content/FPC_model_detected_images.zip')
```
